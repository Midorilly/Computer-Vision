```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

**Exercise 1 - Color spaces**

Take a color image and visualize it. Remember: The image will be read in BGR (Blue-Green-Red), because of OpenCV defaults. OpenCV imshow() function assumes that the image shown is in BGR color-space. Which means, you will always need to convert back to RGB to see what we want

```python
img = cv2.imread('sailor.png')
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2RGB)
```

Resize the image using three different scales (e.g. 0.3, 0.5, 0.7), visualize each rescaled image and for each compute the total number of pixels

```python
img_copy = img_rgb.copy()
scale3 = 0.3
width = int(img.shape[1]*scale3)
height = int(img.shape[0]*scale3)
dimensions=(width, height)
img_resized3 = cv2.resize(img_copy,dimensions)
plt.imshow(img_resized3)
```

```python
img_copy = img_rgb.copy()
scale5 = 0.5
width = int(img.shape[1]*scale5)
height = int(img.shape[0]*scale5)
dimensions=(width, height)
img_resized5 = cv2.resize(img_copy,dimensions)
plt.imshow(img_resized5)
```

```python
img_copy = img_rgb.copy()
scale7 = 0.7
width = int(img.shape[1]*scale7)
height = int(img.shape[0]*scale7)
dimensions=(width, height)
img_resized7 = cv2.resize(img_copy,dimensions)
plt.imshow(img_resized7)
```

Visualize the image using the color spaces: RGB, HSV and Lab

```python
plt.imshow(img_rgb)
```

```python
img_hsv = cv2.cvtColor(img_copy, cv2.COLOR_RGB2HSV)
plt.imshow(img_hsv)
```

```python
img_lab = cv2.cvtColor(img_copy, cv2.COLOR_RGB2Lab)
plt.imshow(img_lab)
```

Split the RGB image into three channels and show separately each channel.

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(20,20))
for i in range(0, 3):
  ax = axs[i]
  ax.imshow(img_rgb[:, :, i], cmap='gray')
axs[0].set_title('RED')
axs[1].set_title('GREEN')
axs[2].set_title('BLUE')
plt.show()
```

**Exercise 2 - Drawing**

Take a color image and show it

```python
img = cv2.imread('sailor.png')
plt.imshow(img)
plt.show()
img_rgb = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

Put a grid 3x3 over the image by drawing vertical and horizontal lines in green color

```python
img_grid = img_rgb.copy()
cv2.line(img_grid, (100,300), (400,300), (0,255,0), 10)
cv2.line(img_grid, (100,400), (400,400), (0,255,0), 10)
cv2.line(img_grid, (100,500), (400,500), (0,255,0), 10)
cv2.line(img_grid, (100,600), (400,600), (0,255,0), 10)

cv2.line(img_grid, (100,300), (100,600), (0,255,0), 10)
cv2.line(img_grid, (200,300), (200,600), (0,255,0), 10)
cv2.line(img_grid, (300,300), (300,600), (0,255,0), 10)
cv2.line(img_grid, (400,300), (400,600), (0,255,0), 10)

plt.imshow(img_grid)
plt.show()
```

Consider the central pixel of your image and draw four arrows of different colors. Each arrow starts in a corner of the image and ends in the central pixel

```python
img_arrow = img_rgb.copy()
dim = img_arrow.shape
center = (int(dim[1]/2), int(dim[0]/2))

cv2.arrowedLine(img_arrow, (0,0), center, (50,100,70), 10)
cv2.arrowedLine(img_arrow, (dim[1],0), center, (200,25,0), 10)
cv2.arrowedLine(img_arrow, (0,dim[0]), center, (34,87,167), 10)
cv2.arrowedLine(img_arrow, (dim[1],dim[0]), center, (98,0,47), 10)

plt.imshow(img_arrow)
plt.show()
```

Create an empty black image and put your name on it using the font and the color your prefer
